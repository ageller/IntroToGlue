## [NEO Earth Close Approaches archive](https://cneos.jpl.nasa.gov/ca/)

```python
neoCA = pd.read_csv('cneos_closeapproach_data.csv')
neoCA
```

### Fix the ID, date, distance and diameter

Also, take only the first occurance of each close approach

```python
from dateutil import parser

pdes = []
date = []
nomDist = []
minDist = []
diameter = []
indices = []
for index, row in neoCA.iterrows():
    #ID
    x = row['Object']
    if (x[0] == '('):
        p1 = x.find(')')
        iden = x[1:p1]
    else:
        p1 = x.find(' ')
        iden = x[0:p1]
        
    #take only the first entry
    if (iden not in pdes):
        indices.append(index)
        pdes.append(iden)
        
        #check to see if this is also listed in the JPL file
        check = neo.loc[neo['pdes'] == pdes[-1]]
        if (len(check) == 0 or len(check) > 1):
            print(pdes[-1], len(check))

        #date
        x = row['Close-Approach (CA) Date']
        p1 = x.find('±') -1
        dt = parser.parse(x[0:p1])
        date.append(dt.year + dt.month/12. + dt.day/365.24 + dt.hour/8760. + dt.minute/(8760.*60.) + dt.second/(8760.*60.*60.))

        #distance in units of au
        x = row['CA Distance Nominal (LD | au)']
        p1 = x.find('|')+1
        nomDist.append(x[p1:])     

        x = row['CA Distance Nominal (LD | au)']
        p1 = x.find('|')+1
        minDist.append(x[p1:])  
        
        #diameter, use the unit of km
        x = row['Diameter']
        if (pd.isna(x)):
            diameter.append(x)
        else:
            xx = x.split()
            if (len(xx) == 5):
                #format '250 m - 570 m'; take the mean
                conv = 1.
                if (xx[1] == 'm'):
                    conv = 1000.
                diam1 = float(xx[0])*conv
                conv = 1.
                if (xx[-1] == 'm'):
                    conv = 1000.
                diam2 = float(xx[-2])*conv
                diameter.append((diam1 + diam2)/2.)
            else:
                p1 = x.find('±')
                if (p1 > 0):
                    # format '0.459±0.004 km'
                    conv = 1.
                    if (xx[1] == 'm'):
                        conv = 1000.
                    diameter.append(float(xx[0][0:(p1-1)])*conv)           
                else:
                    #format '0.33 km'
                    conv = 1.
                    if (xx[1] == 'm'):
                        conv = 1000.
                    diameter.append(float(xx[0])*conv)   
```

### Update the df with the new values

```python
df = neoCA.iloc[indices].copy()
df.insert(0,'pdes',pdes)
df['Close-Approach (CA) Date'] = date
df['CA Distance Nominal (LD | au)'] = nomDist
df['CA Distance Minimum (LD | au)'] = minDist
df['Diameter'] = diameter

df = df.rename(columns={'CA Distance Nominal (LD | au)':'CA Distance Nominal (AU)', 
                        'CA Distance Minimum (LD | au)':'CA Distance Minimum (AU)',
                        'Diameter':'Diameter (km)', 
                        'Unnamed: 8':'extra'})
```

```python
neo_join = neo.join(df.set_index('pdes'), on='pdes')
neo_join
```
